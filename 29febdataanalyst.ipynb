```python
import pandas as pd

# 1️⃣ Monthly Sales
monthly_sales = pd.read_csv('/content/monthly_sales.csv', header=None)  # no header
monthly_sales.columns = ['Year', 'Month', 'TotalUnits', 'TotalBonus', 'TotalDiscount', 'TotalSales']

# 2️⃣ Top Products
top_products = pd.read_csv('/content/top_products.csv', header=None)
top_products.columns = ['ProductName', 'UnitsSold', 'Revenue']

# 3️⃣ Distributor Performance
distributor_perf = pd.read_csv('/content/distributor_performance.csv', header=None)
distributor_perf.columns = ['DistributorName', 'Revenue']

# 4️⃣ Client Type Analysis
client_type = pd.read_csv('/content/client_type_analysis.csv', header=None)
client_type.columns = ['ClientType', 'Clients', 'Revenue']

# 5️⃣ Bonus & Discount Analysis
bonus_discount = pd.read_csv('/content/bonus discounts impact.csv', header=None)
bonus_discount.columns = ['Year', 'Month', 'TotalBonus', 'TotalDiscount']
```

```python
# Fill nulls with 0
for df in [monthly_sales, top_products, distributor_perf, client_type, bonus_discount]:
    df.fillna(0, inplace=True)
```

```python
monthly_sales.to_csv('monthly_sales_clean.csv', index=False)
top_products.to_csv('top_products_clean.csv', index=False)
distributor_perf.to_csv('distributor_perf_clean.csv', index=False)
client_type.to_csv('client_type_clean.csv', index=False)
bonus_discount.to_csv('bonus_discount_clean.csv', index=False)
```
